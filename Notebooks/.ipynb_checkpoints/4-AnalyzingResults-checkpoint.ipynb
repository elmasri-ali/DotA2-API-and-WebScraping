## Creating a Database

```python
# Load datasets
chats_df = pd.read_csv('Chats.csv')
hero_merged_df= pd.read_csv('Hero_merged.csv')
matches_df = pd.read_csv('Matches.csv')
teamspirit_cleaned_df = pd.read_csv('TeamspiritCleaned.csv')
```

```python
# Connect to SQLite database 
conn = sqlite3.connect('dota2.db')
cursor = conn.cursor()
```

```python
# Load each dataset into the database as individual tables
chats_df.to_sql('Chats', conn, if_exists='replace', index=False)
hero_merged_df.to_sql('Hero_merged', conn, if_exists='replace', index=False)
matches_df.to_sql('Matches', conn, if_exists='replace', index=False)
teamspirit_cleaned_df.to_sql('TeamspiritCleaned', conn, if_exists='replace', index=False)
```

### Dataframe Querying 

```python
# Execute the SQL query
query = """
SELECT Matches.*,
       TeamspiritCleaned.position,
       Chats.type,
       Chats.key,
       Hero_merged.*
FROM Matches
JOIN TeamspiritCleaned ON Matches.match_id = TeamspiritCleaned.Match_ID AND Matches.Account_id = TeamspiritCleaned.Account_id
JOIN Chats ON Matches.match_id = Chats.match_id AND Matches.Player_slot = Chats.Player_slot
JOIN Hero_merged ON Matches.Hero_id = Hero_merged.Id
"""
cursor.execute(query)

# Fetch the results
results = cursor.fetchall()
```

```python
df = pd.DataFrame(results, columns=[column[0] for column in cursor.description])
```

```python
conn.close()
```

## Visualization 


```python
# Set the default figure size for all plots
plt.rcParams['figure.figsize'] = [10, 6]  # Width, height in inches
```

### 1- Distribution of Match Durations

```python
sns.histplot(df['Duration'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Match Durations')
plt.xlabel('Duration (in minutes)')
plt.ylabel('Frequency')
plt.show()
```

**This histogram showcases the distribution of match durations in minutes for the Dota 2 matches involving Team Spirit. The use of a kernel density estimate (KDE) overlay provides a smooth representation of the distribution, highlighting the most common match lengths. The visualization suggests a certain range of game durations that occur more frequently, which could indicate typical game lengths or popular strategies that result in matches of these durations.**

### 2- Avergae Kills, Deaths and Assists per Hero

```python
#Grouping and Aggregation
grouped_data = df.groupby('Hero').agg(
    Win_Rate=pd.NamedAgg(column='Win', aggfunc=lambda x: x.mean() * 100),
    Average_Kills=pd.NamedAgg(column='Kills', aggfunc='mean'),
    Average_Deaths=pd.NamedAgg(column='Deaths', aggfunc='mean'),
    Average_Assists=pd.NamedAgg(column='Assists', aggfunc='mean')
).reset_index()

grouped_data_sorted = grouped_data.sort_values(by='Win_Rate', ascending=False)
```

```python
# Creating subplots for Average Kills, Deaths, and Assists per Hero

fig, axes = plt.subplots(1, 3, figsize=(20, 10))

# Average Kills per Hero
top_kills = grouped_data_sorted.sort_values('Average_Kills', ascending=False).head(10)
sns.barplot(ax=axes[0], x='Hero', y='Average_Kills', data=top_kills, palette='viridis')
axes[0].set_title('Top 10 Heroes by Average Kills')
axes[0].set_xlabel('Hero ID')
axes[0].set_ylabel('Average Kills')
axes[0].tick_params(axis='x', rotation=45)

# Average Deaths per Hero
top_deaths = grouped_data_sorted.sort_values('Average_Deaths', ascending=False).head(10)
sns.barplot(ax=axes[1], x='Hero', y='Average_Deaths', data=top_deaths, palette='inferno')
axes[1].set_title('Top 10 Heroes by Average Deaths')
axes[1].set_xlabel('Hero ID')
axes[1].set_ylabel('Average Deaths')
axes[1].tick_params(axis='x', rotation=45)

# Average Assists per Hero
top_assists = grouped_data_sorted.sort_values('Average_Assists', ascending=False).head(10)
sns.barplot(ax=axes[2], x='Hero', y='Average_Assists', data=top_assists, palette='cubehelix')
axes[2].set_title('Top 10 Heroes by Average Assists')
axes[2].set_xlabel('Hero ID')
axes[2].set_ylabel('Average Assists')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**This set of bar plots presents the top 10 heroes in terms of average kills, deaths, and assists, based on the aggregated data from Team Spirit's matches. Each subplot focuses on one of the metrics, offering insights into which heroes tend to have higher involvement in kills, which are more likely to die, and which contribute to kills through assists. This visualization aids in understanding hero dynamics and their roles in matches.**

### 3-Average KDA by Position for Team Spirit Players

```python
# Filter for Team Spirit players
team_spirit_players = df[df['Is_teamspirit'] == True]

# Calculate average KDA by Position for Team Spirit players
average_kda_by_position = team_spirit_players.groupby('Position')['Kda'].mean().reset_index()

# Visualization of Average KDA by Player Position for Team Spirit Players

sns.barplot(x='Position', y='Kda', data=average_kda_by_position, palette='coolwarm')
plt.title('Average KDA by Player Position for Team Spirit')
plt.xlabel('Player Position')
plt.ylabel('Average KDA')
plt.xticks(rotation=45)  # Rotate tick labels for better readability
plt.show()
```

**The bar plot visualizes the average Kill-Death-Assist (KDA) ratio for Team Spirit players, segmented by their playing positions (from pos1 to pos5). This analysis highlights the performance efficiency of players in different roles, providing insights into how each position contributes to the team's overall success. Positions with higher KDA ratios indicate roles where players are particularly effective in contributing to kills while avoiding deaths.**

### 4-KDA and Chat Volume Relationship by Position

```python
# Count the number of chats per Match_id and Account_id combination
chat_count_per_game = team_spirit_players.groupby(['Match_id', 'Account_id']).size().reset_index(name='Chat_Count')

# Merge this count back with the Team Spirit 
team_spirit_with_chat_count = pd.merge(team_spirit_players, chat_count_per_game, on=['Match_id', 'Account_id'], how='left')
```

```python
# Create a subplot with 5 graphs for each position.
fig, axes = plt.subplots(5, 1, figsize=(10, 20), sharex=False)

# List of unique positions
positions = team_spirit_with_chat_count['Position'].unique()

for i, position in enumerate(positions):
    # Filter data for the current position
    position_data = team_spirit_with_chat_count[team_spirit_with_chat_count['Position'] == position]
    
    # Plot
    sns.regplot(ax=axes[i], x='Chat_Count', y='Kda', data=position_data, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
    axes[i].set_title(f'Position {position} - Chat Volume (Count) and KDA Relationship')
    axes[i].set_xlabel('Total Chat Count per Game')
    axes[i].set_ylabel('KDA')

plt.tight_layout()
plt.show()
```

**Through a series of scatter plots with regression lines, this visualization explores the relationship between the total chat volume per game and the KDA ratio for Team Spirit players, segmented by position. The plots could reveal how communication (as approximated by chat volume) correlates with performance metrics like KDA. A positive correlation might suggest that more in-game communication is associated with better performance, depending on the role.**

### 5-Win rate by Hero Attack Type by Position

```python
# Setting up the subplots
fig, axes = plt.subplots(len(positions), 1, figsize=(10, 18), sharex=True)

for i, position in enumerate(positions):
    # Filter data for the current position
    position_data = team_spirit_with_chat_count[team_spirit_with_chat_count['Position'] == position]
    
    # Calculate win rate by Attack Type for the current position
    win_rate_by_attack_type = position_data.groupby('Attack Type').agg(
        Win_Rate=pd.NamedAgg(column='Win', aggfunc=lambda x: x.mean() * 100)
    ).reset_index()
    
    # Plotting the win rate by Attack Type for the current position
    sns.barplot(ax=axes[i], x='Attack Type', y='Win_Rate', data=win_rate_by_attack_type, palette='muted')
    axes[i].set_title(f'Win Rate by Attack Type for {position}')
    axes[i].set_xlabel('Attack Type')
    axes[i].set_ylabel('Win Rate (%)')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate tick labels for better readability

plt.tight_layout()
plt.show()
```

**This series of bar plots examines the win rates associated with different hero attack types (melee or ranged), further broken down by player positions. By comparing the win rates across positions for different attack types, this analysis can uncover trends or preferences in hero selection that correlate with higher chances of winning. It may also highlight strategic choices in hero selection based on positions that contribute to the team's success.**

#### Nested Pie Chart

```python
# Aggregate Win Rate by Attack Type for All Positions
all_positions_data = df.groupby(['Position', 'Attack Type']).agg(
    Win_Rate=pd.NamedAgg(column='Win', aggfunc=lambda x: x.mean() * 100)
).reset_index()
```

```python
outer_values = [all_positions_data[all_positions_data['Position'] == position]['Win_Rate'].sum() for position in positions]

# Inner Slices: Win Rate by Attack Type within each Position
inner_values = []
inner_labels = []
for position in positions:
    position_data = all_positions_data[all_positions_data['Position'] == position]
    for _, row in position_data.iterrows():
        inner_values.append(row['Win_Rate'])
        inner_labels.append(f"{row['Attack Type']} ({row['Position']})")

fig, ax = plt.subplots()
size = 0.3
cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10, 13, 14, 17, 18]))

# Creating the outer ring
outer_pies, _ = ax.pie(outer_values, radius=1, colors=outer_colors,
                       wedgeprops=dict(width=size, edgecolor='w'), labeldistance=1.05)

# Creating the inner ring
inner_pies, x, _ = ax.pie(inner_values, radius=1-size, colors=inner_colors, 
                       labels=inner_labels, 
                       wedgeprops=dict(width=size, edgecolor='w'), autopct='%1.1f%%')



plt.show()
```

**The nested pie chart similarly to the subplots bar chart illustrates win rates for Melee and Ranged heroes across the five positions in Dota 2. The outer ring's segment size depicts each position's win contribution, while the inner slices show the win rates for each attack type within those positions. This visualization helps identify successful hero types per position, potentially guiding hero selection strategies for future matches. The chart's clear labels and color coding facilitate quick assessment of performance trends across positions.**
