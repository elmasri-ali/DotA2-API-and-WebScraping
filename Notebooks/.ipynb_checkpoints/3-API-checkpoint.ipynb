### Step1: Function to fetch matches based om Match_ID from API.

```python
def fetch_matches_from_csv(csv_file_path):
     
    """
    Fetches detailed data for each match specified in the CSV file.

    This function reads match IDs from the specified CSV file and uses these IDs to
    make HTTP GET requests to the opendota predefined BASE_URL to retrieve data for each match.
    It respects API rate limits by introducing a delay between consecutive requests.
    """
        
        
    matches_df = pd.read_csv(csv_file_path)
    match_ids = matches_df['Match_ID'].tolist()

    matches_data = []

    for match_id in match_ids:
        print(f"Fetching data for match ID: {match_id}...")
        response = requests.get(f"{BASE_URL}/matches/{match_id}?api_key={API_KEY}")
        if response.status_code == 200:
            match_data = response.json()
            matches_data.append(match_data)
        else:
            print(f"Failed to fetch data for match ID: {match_id}")
       
        # Introduce a delay to avoid hitting rate limits
        sleep(0.5)

    return matches_data
```

```python
matches_data = fetch_matches_from_csv(csv_path)
```

The returned JSON response is a complex list of nested dictionaries with multiple nesting levels. We will explore it further. 

### Step2: Function to explore the complexity of the fetched JSON data.

```python
# Get Max depth for the json reponse and count of elements to explore the complexity of the data. 
def count_elements_at_each_level(obj, level=0, counts=None):
    if counts is None:
        counts = {}
    if isinstance(obj, dict):
        counts[level] = counts.get(level, 0) + len(obj)
        for value in obj.values():
            count_elements_at_each_level(value, level + 1, counts)
    elif isinstance(obj, list):
        counts[level] = counts.get(level, 0) + len(obj)
        for item in obj:  # Corrected line: Iterate over each item in the list
            count_elements_at_each_level(item, level + 1, counts)
    return counts
```

```python
counts = count_elements_at_each_level(matches_data)
print(counts)
```

### Step3: Extracting nested chat data into a data frame.

```python
# Extract chat lists, AND match ID
chat_lists = [
    {**chat, 'match_id': match['match_id']}
    for match in matches_data if 'chat' in match
    for chat in match['chat']
]

# Convert the list of dictionaries to a pandas DataFrame
chat_lists_df = pd.DataFrame(chat_lists)
```

```python
chat_lists_df.tail()
```

### Step4: Extracting Valuable Columns into a new Data frame.

```python
# Extracting the relvant columns or information into a panda dataframe from the nested dictionaries. 
rows = []
for match in matches_data:
    match_id = match["match_id"]
    start_time = match["start_time"]
    for player in match["players"]:
        account_id = player["account_id"]
        row = {
            "account_id": account_id,
            "match_id": match_id,
            "start_time": start_time,
            "player_slot": player.get('player_slot'),
            "duration": player.get("duration"),
            "isRadiant": player.get("isRadiant"),
            "win": player.get("win"),
            "hero_id": player.get("hero_id"),
            "kills": player.get("kills"),
            "deaths": player.get("deaths"),
            "assists": player.get("assists"),
            "kda": player.get("kda"),
            "last_hits": player.get("last_hits"),
            "denies": player.get("denies"),
            "net_worth": player.get("net_worth"),
            "total_gold": player.get("total_gold"),
            "gold_per_min": player.get("gold_per_min"),
            "total_xp": player.get("total_xp"),
            "hero_damage": player.get("hero_damage"),
            "abandon": player.get("abandons")
        }
        rows.append(row)
```

### Step5: Set Index.

```python
# Create DataFrame
player_df = pd.DataFrame(rows)

# Set multi-level index
player_df.set_index(["match_id","account_id"], inplace=True)
```

 ### Step6: Creating new features.

```python
# New column to Indicate if players are of Team Spirit
player_df['Is_TeamSpirit'] = player_df.index.get_level_values('account_id').isin(df_csv['Account_ID'])
```

```python
player_df['gold_lost'] = player_df['total_gold'] - player_df['net_worth']
```

### Step7: Data Transformation.

```python
player_df['duration'] = player_df['duration'] / 60
```

**Transform the Start time column into a panda datetime object.**

```python
player_df['start_time'] = pd.to_datetime(player_df['start_time'], unit='s')  # unit='s' is for Unix timestamp (seconds)
```
